```python
from visTools_v2.core_functions.robust_line import *
```

```python
im = load_img('img/2viewp.jpg')
im1 = canny(sharp=True,
             blur=True,h=3,w=3,sig=0.5,th=.1,Red=0)
im1.fit(im,nonMaxSup=False)
im1.plot()
im1.vector_field(figsize=(10,16),tan=True)
```

```python
RANSAC_par = {
    'RANSAC_P':.99,
    'RANSAC_p':.1,
    'RANSAC_minInliers':100,
    'RANSAC_th':30,
    'RANSAC_lamb':50,
    'RANSAC_ratio':(1,1),
    }

model = RobustFittingLine(sig=1,nIter=20,**RANSAC_par)
model.fit(im1)
model.plot()
```
