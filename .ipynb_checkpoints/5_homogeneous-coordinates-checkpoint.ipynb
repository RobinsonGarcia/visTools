```python
class point:
    def __init__(self,xks,type_='cart'):
        self.input_type = type_
        self.xks = xks
        self.nKps = xks.shape[1]
    '''returns cartesian coordinates'''
    def c(self):
        if self.input_type=='cart':
            return self.xks
        else:
            return self.xks[:2,:]/self.xks[2,:]
    
    '''returns homogeneous coordinates'''
    def h(self):
        if self.input_type=='cart':
            return np.vstack((self.xks,np.ones(self.nKps)))
        else:
            return self.xks
```

```python
def get_random_xks(xks,n=2):
    idx = np.arange(xks.shape[1])
    idx = np.random.choice(idx,n)
    xk_list = []
    return xks[:,idx]
```

```python
class line:
    def __init__(self,l=None,xks=None):
        self.l = l
        if isinstance(l,np.ndarray):
            self.f = self.homog(l)
            
    def line_from_hpoints(self,xk1,xk2):
        self.l = np.cross(xk1,xk2)
        self.f = self.homog(self.l)
        self.xmin = min(xk1[0],xk2[0])
        self.xmax = max(xk1[0],xk2[0])
    
    def __mul__(self,other):
        intersect = np.cross(self.l,other.l)
        intersect = intersect[:2]/intersect[2]
        self.intersect = intersect
        other.intersect = intersect
        
        
    def draw(self,xmin=0,xmax=10,n=10):
        xrange = np.linspace(self.xmin,self.xmax,n)
        return xrange,self.f(xrange)
    def homog(self,l):
        a,b,c = l
        return lambda x:-(a/b)*x-c/b
    def plot(self,c='b',xmin=None,xmax=None):
        if xmin!=None:self.xmin=xmin
        if xmax!=None:self.xmax=xmax
        x,y = self.draw(self.xmin,self.xmax,n=10)
        plt.plot(x,y,c=c)
        
```

```python
def plot_circles(circles,n=5):
    
    for i in range(n):
        try:
            circle = plt.Circle((np.array(circles[-i])[0],np.array(circles[-i])[1]),np.array(circles[-i])[2],fill=False)
            #circle = plt.Circle((10,10),i,fill=False)
            ax=plt.gca()
            ax.add_patch(circle)
            plt.axis('scaled')
        except:
            print('less than  {} candidates'.format(n))
    plt.show()

```

```python
import pylab as pl
from IPython import display as dp
import time

def get_proposals(n_trials,x,y,n,th=10,min_inliers=100):
    circles = []
    N = x.shape[0]
    for i in range(n_trials):
        
        c = -n*x-y

        ls = np.vstack((n,np.ones(n.shape),c))
        
        N1 = np.random.randint(N)
        N2 = np.random.randint(N)

        l1,l2,pts1,pts2 = get_pairs(ls,x,y)
        #l1 = ls[:,N1]
        #l2 = ls[:,N2]
        l1 = line(l=l1)
        l2 = line(l=l2)
        l1*l2
        
        xc,yc = l1.intersect
        
        for R in np.random.permutation(np.arange(5,10,1)):
            R_sq = (x - xc)**2+(y-yc)**2
            e = np.abs(R-np.sqrt(R_sq))
            n_inliers = np.sum(e<th)
            if n_inliers>min_inliers:
                circles.append((xc,yc,R,n_inliers))
                break
                
    return circles
```

```python
def quad(teta):
    if (teta>=0)&(teta<np.pi/2):
        return 0,0
    if (teta>=np.pi/2)&(teta<np.pi):
        return 1,1
    if (teta>=np.pi)&(teta<2*np.pi/3):
        return 2,2
    if (teta>2*np.pi/3)&(teta<2*np.pi):
        return 3,3
```

```python
def get_pairs(ls,x,y):
    f = np.vectorize(quad)
    N = len(ls[0,:])
    N1 = np.random.randint(N)
    pts1 = x[N1],y[N1]
    
    a,b,c = ls[:,N1]
    N1_opo = f(a)[0]
    N2_quad = f(ls[0,:])[0]

    candidates = N1_opo==N2_quad
    n_candidates = np.sum(candidates)
    N2 = np.random.randint(n_candidates)
    pts2 = x[N1_opo==N2_quad][N2],y[N1_opo==N2_quad][N2]


    return ls[:,N1],ls[:,N1_opo==N2_quad][:,N2],pts1,pts2
```

```python
mask_i = np.arange(len(data['kps']))[1:]
for i in mask_i:
    x,y = data['kps'][i]
    n = data['rhos'][i]

    circles = get_proposals(n_trials=1000,x=x,y=y,n=n,th=2.5
                ,min_inliers=50)
    plt.scatter(x,y)
    plot_circles(circles,1)
```
