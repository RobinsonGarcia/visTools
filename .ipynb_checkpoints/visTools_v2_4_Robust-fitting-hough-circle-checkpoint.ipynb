```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
import pickle
f = open('data/coin_data/edgels.pickle','rb')
data = pickle.load(f)
f.close()

xks = np.argwhere(data).T
x,y = xks
```

```python
class HoughTransform:
    def __init__(self,**params):
        self.th = 0.5
        self.Rmin = 1
        self.Rmax = 30
        self.Rdiv = 20
        self.nit=1
        
        self.xx = None
        self.yy = None
        self.R = None
        self.y = None
        self.x = None
        self.N = None
        
        if isinstance(params,dict):
            for k,v in params.items():
                self.__dict__[k]=v
    
    def fit(self,xks):
        x,y = xks
        R = np.linspace(self.Rmin,self.Rmax,self.Rdiv)
        xx,yy = np.meshgrid(np.arange(x.min(),x.max()),np.arange(y.min(),y.max()))
        h1,h2 = xx.shape
        h3 = R.shape[0]
        
        hough = np.zeros((h1,h2,h3))
        for t in range(h1):
            for tt in range(h2):
                xc = np.array([xx[t,tt],yy[t,tt]])
                for ttt,r in enumerate(R):
                    rr = np.sqrt(np.sum((xks - xc[:,np.newaxis])**2,axis=0))
                    errors = np.abs(r - rr)
                    hough[t,tt,ttt] = np.sum(errors<self.th) 
                    
        self.hough = hough
        self.xx = xx
        self.yy = yy
        self.R = R
        self.y = y
        self.x = x
        self.N = h1*h2*h3
        
    
    def get_circle(self,i,plot=False):
        ix,iy,ir = np.unravel_index(np.argsort(self.hough, axis=None)[-i], self.hough.shape)
        xc_ = np.array([self.xx[ix,iy],self.yy[ix,iy]])
        if plot==True:
            
            plt.scatter(self.y,self.x)
            plot_circle_from_cart(xc_,self.R[ir],c='r')
        
        return circle_params_from_cart(xc_,self.R[ir])
    
    def __iter_(self):
        self.nit=1
        return self
    
    def __next__(self):
        if self.nit>self.N:
            raise StopIteration
            
        self.nit+=1
        return self.get_circle(self.nit)
```

```python
params = {
    'Rmin':1,
    'Rmax':20,
    'Rdiv':15,
    'th':0.5    
}

hough = HoughTransform(**params)
hough.fit(xks)
```
