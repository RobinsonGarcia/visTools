```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.ma as ma
import math
import time

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
def load_img(img):
    im = plt.imread(img)
    plt.imshow(im)
    plt.show()
    im = np.moveaxis(im,2,0)
    im = np.mean(im,axis=0)
    im = im[np.newaxis,np.newaxis,:,:]
    return im
```

```python
im = load_img('img/blackline.png')
im1 = canny(sharp=False,
             blur=True,h=3,w=3,sig=1.5,th=0.003,Red=0,g=kernel().prewitt)
im1.fit(im,nonMaxSup=True)
im1.plot()
im1.vector_field()
im1.vector_field(tan=True)
```
