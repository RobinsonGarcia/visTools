```python
def FundamentalM(pairs,min_pts,norm=False):

    Hnorm = HartleysNorm(pairs)
    
    pairs = Hnorm.norm_pairs
    
    idx = np.arange(pairs[0].shape[0])
    idx1 = np.random.choice(idx,min_pts)

    x1 = pairs[0][idx1,0]
    y1 = pairs[0][idx1,1]

    x2 = pairs[1][idx1,0]
    y2 = pairs[1][idx1,1]


    A = np.vstack((x2*x1,x2*y1,x2,y2*x1,y2*y1,y2,x1,y1,np.ones(x2.shape)))

    U,S,V = np.linalg.svd(A.T)

    F_ = V[-1].reshape((3,3))

    U,S,V = np.linalg.svd(F_)
    
    S[-1]=0
    
    F = (U.dot(np.diag(S))).dot(V)
    
    F = Hnorm.denormalize(F)
    
    pairs = Hnorm.pairs
  
    e = F_geomError(pairs,F)


    
    return F,e,pairs[0][idx1,:],pairs[1][idx1,:]
```

```python
def hmg(x):
    n = x.shape[0]
    ones = np.ones(n)
    return np.vstack((x.T,ones)).T
```

```python
def F_geomError(pairs,F):
    e1 = np.abs((np.sum((F.dot(hmg(pairs[1][:,:2]).T))*(hmg(pairs[0][:,:2]).T),axis=0)))
    
    e2 = np.abs((np.sum((((hmg(pairs[0][:,:2])).dot(F)))*(hmg(pairs[1][:,:2])),axis=1)))
       
    return (e1+e2).T

#e1 = (F.dot(hmg(pairs[1][:,:2]).T)).T
#e2 = ((hmg(pairs[0][:,:2])).dot(F))
```

```python

'''
p - probability of being an inlier
k - number of samples
S - number of required trials
'''
def RANSAC2(pairs,min_pts=4,p=0.1,P=.99,th=100,type_='affine',max_it=50000):
    print("Starting RANSAC:")
    #pairs = (pairs[1],pairs[0])
    S = np.round(np.log(1-P)/(np.log(1-p**min_pts))).astype(int)

    best_fit=0

    for i in range(S):
        if p>0.9:
            break
            
        if type_=='affine':
            w,e,x1,x2 = Affine(pairs,min_pts)
        if type_ == 'F':
            w,e,x1,x2 = FundamentalM(pairs,min_pts,norm=False)
        else:
            w,e,x1,x2 = Homography(pairs,min_pts)

        inliers = np.sum(e<th**2)

        if best_fit<inliers:
            best_w = w
            best_e = np.sum(e)

            best_x1 = x1
            best_x2 = x2
            idx = e<th**2
            best_fit = inliers
            p = best_fit/pairs[0].shape[0]
            print('Iteration: {}'.format(i))
            print(best_e,p,best_fit,pairs[0].shape[0])
          
            S += np.round(np.log(1-P)/(np.log(1-p**min_pts))).astype(int)
        if i>max_it:
            break



    if type_=='affine':
        best_w,e,x1,x2 = Affine((pairs[0][idx],pairs[1][idx]),len(pairs[0][idx]))
        e = affine_error(pairs,best_w)
    elif type_=='F':
        best_w,e,x1,x2 = FundamentalM((pairs[0][idx],pairs[1][idx]),len(pairs[0][idx]))
        e = F_geomError(pairs,best_w)
    else:
        best_w,e,x1,x2 = Homography((pairs[0][idx],pairs[1][idx]),len(pairs[0][idx]))
        e = homography_error(pairs,best_w)

    best_x1 = x1
    best_x2 = x2
    print('RANSAC error: {}'.format(e))

    if type_=='affine':
        H = np.reshape(best_w,(2,3))
    elif type_=='F':
        H = np.reshape(best_w,(3,3))
    else:
        H = np.reshape(best_w,(3,3))

    return H,idx
```

```python
class HartleysNorm:
    def __init__(self,pairs):
        self.pairs=pairs
        self.Tl = get_T(hmg(pairs[0][:,:2]))
        self.Tr = get_T(hmg(pairs[1][:,:2]))
        self.norm_pairs = hmg(pairs[0][:,:2]).dot(self.Tl),hmg(pairs[1][:,:2]).dot(self.Tr)
        
    def denormalize(self,F):
        return ((self.Tr.T).dot(F)).dot(self.Tl)
        
```

```python
def get_epipoles(F):
    U,S,V = np.linalg.svd(F)
    el = U[-1]
    er = V[-1]
    return el,er
```

```python
def computeE(el):
    zeros = np.zeros((3,3))
    e = np.array([[0,-el[2],el[1]],[el[2],0,-el[0]],[-el[1],el[0],0]])
    e1 = np.hstack([e,zeros,zeros])
    e2 = np.hstack([zeros,e,zeros])
    e3 = np.hstack([zeros,zeros,e])
    return np.vstack((e1,e2,e3))

```

```python
def computeA(pairs):
    x1 = pairs[0][:,0]
    y1 = pairs[0][:,1]

    x2 = pairs[1][:,0]
    y2 = pairs[1][:,1]


    return np.vstack((x2*x1,x2*y1,x2,y2*x1,y2*y1,y2,x1,y1,np.ones(x2.shape))).T
```

```python
def AlgError(el):
    r = 8
    A = computeA(pairs)
    
    E = computeE(el)

    U,S,V = np.linalg.svd(E)

    U_ = U[:,:r]

    _,_,V1 = np.linalg.svd(A.dot(U_))
    x_ = V1[-1]

    F_ = (U_.dot(x_))

    return np.sum(np.abs(A.dot(F_)))
```

```python
def Jacobian(fun,x0,step):
    delta = np.zeros(x0.shape[0])    
    grad = np.zeros(9,9)
    for i in np.arange(x0.shape[0]):
        for j in np.arange(x0.shape[0])
            delta[i] = step
            x2 = x0 + delta
            x1 = x0 - delta
            increment2 = fun(x2)
            increment1 = fun(x1)
            grad[i,j]=(increment2-increment1)/(2*step)
            delta[i] -= step  

    return np.array(grad)
```

```python
def get_T(data):
    ml = data
    
    n = data.shape[0]

    ml_ = np.mean(ml,axis=0)

    sl = np.sqrt(np.sum((ml - ml_)**2)/(2*n))
   
    return np.array([[1/sl,0,(-1/sl)*ml_[0]],[0,1/sl,(-1/sl)*ml_[1]],[0,0,1]])
```

```python
class Data_norm:
    def __init__(self,pairs):

        n = pairs[0].shape[0]
        mlx = pairs[0][:,0].mean()
        mly = pairs[0][:,1].mean()

        mrx = pairs[1][:,0].mean()
        mry = pairs[1][:,1].mean()

        dl = pairs[0][:,:2] - (mlx,mly)
        dl = np.sqrt(np.mean(np.sum(dl**2,axis=1))/2)

        dr = pairs[1][:,:2] - (mrx,mry)
        dr = np.sqrt(np.mean(np.sum(dr**2,axis=1))/2)


        Hl = np.array([[1/dl,0,-mlx/dl],[0,1/dl,mly],[0,0,1]])

        self.Hl_inv = np.linalg.inv(Hl)

        Hr = np.array([[1/dr,0,-mrx/dr],[0,1/dr,mry],[0,0,1]])
        self.Hr_inv = np.linalg.inv(Hr)

        Xl = np.vstack((pairs[0][:,:2].T,np.ones_like(pairs[0][:,1])))
        Xr = np.vstack((pairs[1][:,:2].T,np.ones_like(pairs[1][:,1])))

        self.Xl = (Hl.dot(Xl)).T
        self.Xl = self.Xl[:,:2]/self.Xl[:,2][:,np.newaxis]

        self.Xr = (Hr.dot(Xr)).T
        self.Xr = self.Xr[:,:2]/self.Xr[:,2][:,np.newaxis]
        self.pairs = self.Xr,self.Xl


    def invert_F(self,F):
        return (self.Hl_inv.dot(F)).dot(self.Hr_inv)
```
